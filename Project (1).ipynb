```python
from matplotlib import image
from matplotlib import pyplot
img1 = image.imread('RPS.png')
pyplot.imshow(img1)
pyplot.axis('off')
pyplot.show()

import random
print("Lets play Jankenpon! (***Rock,***Paper,***Scissors)\n")
human_score=0
computer_score=0
 
def get_choices():
  human_choice = input("Enter a choice (\n'r'for ðŸ‘ŠðŸ¾, \n'p' for paperâœ‹ , \n's' for scissorsâœŒï¸): > ").lower()
  options = ["r", "p", "s"]
  computer_choice = random.choice(options)
  choices = {"human": human_choice, "computer": computer_choice}
  return choices 

def check_win(human, computer):
  global human_score
  global computer_score
  print(f"*Your chose: > {human}\n*Computer chose : > {computer}\n")    
  if human == computer:
     return ("It's a tieðŸ¤!\n")
  elif (human == "r"and computer == "s"):
      human_score = human_score+1
      return("Rock smashes scissors! You winðŸ¥‡!\n")
  elif human == "s" and computer == "r":
     computer_score = computer_score+1
     return("Rock smashes scissors! You loseðŸ‘Ž\n")
  elif human == "p"and computer == "r":
     human_score = human_score+1
     return("Paper covers rock! You winðŸ¥‡!\n")
  elif human == "r"and computer == "p":
     computer_score = computer_score+1
     return("Paper covers rock! You loseðŸ‘Ž.\n")
  elif human == "p"and computer == "s":
     computer_score = computer_score+1
     return("Scissors cuts paper! You loseðŸ‘Ž.\n")
  elif human == "s"and computer == "p":
     human_score = human_score+1
     return("Scissors cuts paper! You win ðŸ¥‡!\n")
    
def show_score():
   global human_score 
   global computer_score
   print("Human score:",human_score,"Computer score:",computer_score)
   if human_score > computer_score:
    print("Human won this gameðŸ¤¡")
   elif  human_score <computer_score:
    print("Computer won this time ðŸ¤–")
   elif computer_score == human_score: 
    print("Human and computer tiedðŸ‘»")
   else:
    print("Human or computer!")
    
choices = get_choices()
result = check_win(choices["human"],choices["computer"])
print(result)
show_score()
choice = input("\nDo you want to play again? (Y/N): > ").lower()

while choice =="y":
 choices = get_choices()
 result = check_win(choices["human"],choices["computer"])
 print(result)
 show_score()
 choice = input("\nDo you want to play again? (Y/N): > ").lower()
if choice== "n":
  show_score()
  print("Thanks and Bye")
else:
  print("Please choose given options")  
  
```
